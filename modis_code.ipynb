```python
from urllib.request import urlopen

sensor = 'aqua'
sdate = '2022-01-01'
edate = '2022-01-05'
level = 'L3b'
file_pattern = '*DAY_CHL*'

url = "https://oceandata.sci.gsfc.nasa.gov/api/file_search?sensor={0}&sdate={1}&edate={2}&dtype={3}&search={4}&addurl=1&results_as_file=1" \
               .format(sensor, sdate, edate, level, file_pattern)
print("Requesting", url)  
#print("Requesting", "https://oceandata.sci.gsfc.nasa.gov/api/file_search?sensor=aqua&sdate=2021-12-20&edate=2021-12-25&dtype=L3b&addurl=1&results_as_file=1&search=*DAY_CHL*")
         
resp = urlopen(url)
img_urls = resp.read().decode('utf8').split('\n')

if img_urls[0] != '':
    print("Found {0} files.".format(len(img_urls)))
else:
    print("No files found.")
```
