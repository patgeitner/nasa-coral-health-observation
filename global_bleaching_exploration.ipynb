```python
con = sqlite3.connect("/Users/patrickgeitner/UofR/Capstone/data/Global_Coral_Bleaching_Database_SQLite_11_24_21.db")
cur = con.cursor()

tbls = cur.execute('SELECT name from sqlite_master where type= "table"').fetchall()
print([tbl[0] for tbl in tbls])

sample_event = pd.read_sql_query("SELECT * from Sample_Event_tbl", con)
site_info = pd.read_sql_query("SELECT * from Site_Info_tbl", con)
cover_info = pd.read_sql_query("SELECT * from Cover_tbl", con)
environmental_info = pd.read_sql_query("SELECT * from Environmental_tbl", con)
# bleaching_info = pd.read_sql_query("SELECT * from Bleaching_tbl", con)
```
