```python
site = pd.read_sql_query("SELECT * FROM Site_Info_tbl", cnx)
site
```

```python
event_site_merge = pd.merge(site, event, on='Site_ID', how='outer')
event_site_merge
event_site_merge_concise = event_site_merge[['Latitude_Degrees', 'Longitude_Degrees', 'Date_Year']]
df = event_site_merge_concise[event_site_merge_concise['Date_Year']>=2000]
df

gbr_df = df[((df['Latitude_Degrees']<=-10)&(df['Latitude_Degrees']>=-28.7)&(df['Longitude_Degrees']>=142.69)&(df['Longitude_Degrees']<=153.7)) | (df['Latitude_Degrees']<=27)&(df['Latitude_Degrees']>=17)&(df['Longitude_Degrees']>=-85)&(df['Longitude_Degrees']<=-63)]
gbr_df
```

```python
event_site_merge = event_site_merge[event_site_merge['Date_Year']>=2000]


event_site_merge = event_site_merge[((event_site_merge['Latitude_Degrees']<=-10)&(event_site_merge['Latitude_Degrees']>=-28.7)&(event_site_merge['Longitude_Degrees']>=142.69)&(event_site_merge['Longitude_Degrees']<=153.7)) | (event_site_merge['Latitude_Degrees']<=27)&(event_site_merge['Latitude_Degrees']>=17)&(event_site_merge['Longitude_Degrees']>=-85)&(event_site_merge['Longitude_Degrees']<=-63)]

event_site_bleaching = pd.merge(Bleaching_tbl, event_site_merge, on='Sample_ID', how='inner')
event_site_bleaching
```
